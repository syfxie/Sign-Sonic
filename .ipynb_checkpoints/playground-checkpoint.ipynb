```python
def process_image(image, size=(64, 64)):
    '''
    Returns the resized image raw and as a normalized numpy array
    '''
    resized_image = np.resize(image, size)
    image_array = np.array(resized_image)

    print("Shape of the resized image:", image_array.shape)

    # Noramlize the image array between 0-1
    return image_array / 255.0
```

```python
process_image(img)
```

```python
def process_video(video_path, parent_dest_path, num_frames):
    frames_sequence = []
    resized_image, result = None, None
    
    if os.path.isfile(video_path):
        abs_video_path = os.path.abspath(video_path)
        abs_parent_dest_path = os.path.abspath(parent_dest_path)

        if not os.path.isdir(abs_parent_dest_path):
            os.makedirs(abs_parent_dest_path)

        print("Converting video from: %s into frames at %s" % (abs_video_path, abs_parent_dest_path))

        cap = cv.VideoCapture(abs_video_path)
        frames_collected = 0
        interval = 20
        count = 0
        
        while frames_collected <  num_frames:
            ret, frame = cap.read()  # extract frame
            
            if not ret:
                break

            # process and save every one in twenty frames
            if count % interval == 0:
                frame_path = os.path.join(abs_parent_dest_path, "{}.jpeg".format(str(frames_collected)))
                
                print('Keypoints for video {}, frame {}: '.format(video_path,  frames_collected))
                
                if not os.path.exists(frame_path):
                    resized_image, result = process_image(frame)
                    cv.imwrite(frame_path, resized_image) # save image to file
                    frames_sequence.append(result)
                frames_collected += 1
            
            count += 1

            if cv.waitKey(10) & 0xFF == ord('q'):
                break
        cap.release()

        # repeat last frame until the target number of frames are collected
        while result and resized_image and frames_collected <  num_frames:
            frame_path = os.path.join(abs_parent_dest_path, "{}.jpeg".format(str(frames_collected)))
            # results = extract_image_keypoints(frame)

            if not os.path.exists(frame_path):
                # lastFrame = process_image(lastFrame)
                cv.imwrite(frame_path, resized_image) # save image to file
                frames_sequence.append(result)
            frames_collected += 1

    return frames_sequence
```
